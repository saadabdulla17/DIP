```python
import numpy as np
import matplotlib.pyplot as plt

time = np.linspace(0, 1, 1000)
frequency_carrier = 10
frequency_modulator = 2
amplitude = 1
fm_signal = amplitude * np.sin(2 * np.pi * frequency_carrier * time + np.sin(2 * np.pi * frequency_modulator * time))

plt.plot(time, fm_signal)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('FM Signal (Time Domain)')
plt.show()
```
