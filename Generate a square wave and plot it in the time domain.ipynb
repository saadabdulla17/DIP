```python
import numpy as np
import matplotlib.pyplot as plt

time = np.linspace(0, 1, 1000)
frequency = 5  # Frequency of the square wave
amplitude = 1  # Amplitude of the square wave
duty_cycle = 0.5  # Duty cycle of the square wave
square_wave = amplitude * np.sign(np.sin(2 * np.pi * frequency * time))

plt.plot(time, square_wave)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Square Wave (Time Domain)')
plt.show()
```
