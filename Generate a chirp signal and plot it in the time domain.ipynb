```python
import numpy as np
import matplotlib.pyplot as plt

time = np.linspace(0, 1, 1000)
frequency_start = 1
frequency_end = 10
chirp_signal = np.sin(2 * np.pi * (frequency_start + (frequency_end - frequency_start) * time) * time)

plt.plot(time, chirp_signal)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Chirp Signal (Time Domain)')
plt.show()
```
